```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(5)
```

```python
credits.head(5)
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
movies.info()
```

```python

movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head(2)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[5].genres
```

```python
import ast
def filter_name(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies['genres'].apply(filter_name)
```

```python
movies['keywords']=movies['keywords'].apply(filter_name)
```

```python
movies.head(2)
```

```python
import ast
def filter5_name(obj):
    l=[]
    ct=0;
    for i in ast.literal_eval(obj):
        if ct!=5:
            l.append(i['name'])
            ct+=1
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(filter5_name)
```

```python
movies.head(1)
```

```python
movies['crew'][0]
```

```python
def filter_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew']=movies['crew'].apply(filter_director)
```

```python
movies.head(2)
```

```python
movies['overview'][0]
```

```python
movies['overview']=movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace("  ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace("  ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tag']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
movies.head(2)
```

```python
new_data=movies[['movie_id','title','tag']]
new_data
```

```python
new_data['tag']=new_data['tag'].apply(lambda x:" ".join(x))
new_data
```

```python
new_data['tag'][0]
```

```python
new_data['tag']=new_data['tag'].apply(lambda x:x.lower())
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_data['tag']).toarray()
vectors
```

```python
vectors[0]
```

```python
feature=cv.get_feature_names_out()
print(feature[:600:])
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
print(stem('actions'))
print(stem('activities'))
```

```python
new_data['tag']=new_data['tag'].apply(stem)
```

```python
vectors=cv.fit_transform(new_data['tag']).toarray()
vectors
```

```python
feature=cv.get_feature_names_out()
print(feature[:600:])
```

```python
from sklearn.metrics.pairwise import cosine_similarity as cos_sim
```

```python
similarity=cos_sim(vectors)
similarity[1]
```

```python
def recommend(movie):
    movie_ind=new_data[new_data['title']==movie].index[0]
    dist=similarity[movie_ind]
    movies_list=sorted(list(enumerate(dist)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_data.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
